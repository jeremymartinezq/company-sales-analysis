```python
# Company Sales Analysis

# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Question 1: Load and display the first few rows of the 'company_sales_data.csv'
# ------------------------------------------------------
# Loading company sales data
sales_data = pd.read_csv('company_sales_data.csv')
# Displaying the first few rows
sales_data.head()

# Visualization for Question 1
plt.figure(figsize=(8, 4))
plt.plot(sales_data['month_number'], sales_data['total_profit'], color='blue', marker='o')
plt.title('Question 1: Company Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(sales_data['month_number'])
plt.grid(True, linestyle='--')
plt.show()

# Question 2: Plotting Total Profit and Toothpaste Sales of All Months
# ------------------------------------------------------
# Extracting month, total profit, and toothpaste sales data
months = sales_data['month_number']
total_profit = sales_data['total_profit']
toothpaste_sales = sales_data['toothpaste']

# Plotting total profit with proper formatting
plt.figure(figsize=(10, 5))
plt.plot(months, total_profit, label='Total Profit', color='green', linestyle='--', marker='o')

# Adding total profit numbers above each point
for i, txt in enumerate(total_profit):
    plt.annotate(txt, (months[i], total_profit[i]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color='green')

# Plotting toothpaste sales as scatter plot
plt.scatter(months, toothpaste_sales, label='Toothpaste Sales', color='blue', marker='x')

# Adding total toothpaste sales numbers above each point
for i, txt in enumerate(toothpaste_sales):
    plt.annotate(txt, (months[i], toothpaste_sales[i]), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color='blue')

# Highlighting the month with max profit
max_profit = total_profit.max()
max_profit_month = months[total_profit.idxmax()]
plt.scatter(max_profit_month, max_profit, color='red', s=100, label='Max Profit', edgecolors='black')

# Adding titles, labels, and customizing ticks
plt.title('Question 2: Total Profit and Toothpaste Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales / Profit')
plt.legend(loc='upper left')
plt.xticks(months)
plt.yticks(np.arange(100000, 600001, 100000))
plt.grid(True, linestyle='--')
plt.show()

# Question 3: Display Product Sales Distribution as Pie Chart
# ------------------------------------------------------
# Summing sales data for each product
product_sales = sales_data[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()

# Pie Chart Visualization for Product Sales Distribution
plt.figure(figsize=(8, 8))
product_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Question 3: Product Sales Distribution (Pie Chart)')
plt.ylabel('')
plt.show()

# Question 3b: Plotting Sales Data of All Products Over Months
# ------------------------------------------------------
# Plotting sales data for all products
plt.figure(figsize=(12, 6))
plt.plot(months, sales_data['facecream'], label='Face Cream Sales', marker='o')
plt.plot(months, sales_data['facewash'], label='Face Wash Sales', marker='o')
plt.plot(months, sales_data['toothpaste'], label='Toothpaste Sales', marker='o')
plt.plot(months, sales_data['bathingsoap'], label='Bathing Soap Sales', marker='o')
plt.plot(months, sales_data['shampoo'], label='Shampoo Sales', marker='o')
plt.plot(months, sales_data['moisturizer'], label='Moisturizer Sales', marker='o')

# Adding titles and labels
plt.title('Question 3b: Sales Data of All Products Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(months)
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()

# Question 4: Total Car Count from 'automobile_data.csv'
# ------------------------------------------------------
# Loading automobile data for car count
auto_data = pd.read_csv('automobile_data.csv')

# Counting total cars for each company
company_car_counts = auto_data['company'].value_counts()
print("\nTotal car count for each company:")
print(company_car_counts)

# Visualization of total car counts
plt.figure(figsize=(12, 6))
company_car_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Question 4: Total Car Counts by Company')
plt.xlabel('Company')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()

# Question 5: Numpy Array Manipulation and Car Count from 'automobile_data.csv'
# ------------------------------------------------------
# Creating the original array
original_array = np.array([[34, 43, 73], [82, 22, 12], [53, 94, 66]])
print("Original array:")
print(original_array)

# Deleting the second column
modified_array = np.delete(original_array, 1, axis=1)
print("\nArray after deleting column 2 on axis 1:")
print(modified_array)

# Inserting a new column at the second position
final_array = np.insert(modified_array, 1, 10, axis=1)
print("\nArray after inserting column 2 on axis 1:")
print(final_array)

# Visualization for Question 4 (Array Manipulation)
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
axs[0].imshow(original_array, cmap='Blues')
axs[0].set_title('Question 5: Original Array')

axs[1].imshow(modified_array, cmap='Blues')
axs[1].set_title('Question 5: After Deleting Column')

axs[2].imshow(final_array, cmap='Blues')
axs[2].set_title('Question 5: After Inserting Column')

for ax in axs:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
